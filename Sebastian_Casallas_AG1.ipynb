```python
#Cambio de monedas - Técnica voraz
################################################################################
SISTEMA = [25, 10, 5, 1]
################################################################
def cambio_monedas(CANTIDAD,SISTEMA):
#....
  SOLUCION = [0]*len(SISTEMA)
  ValorAcumulado = 0

  for i,valor in enumerate(SISTEMA):
    monedas = (CANTIDAD-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if CANTIDAD == ValorAcumulado:
      return SOLUCION

  print("No es posible encontrar solucion")

cambio_monedas(37,SISTEMA)


################################################################
```

```python
#N Reinas - Vuelta Atrás()
################################################################################

#Verifica que en la solución parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solución al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)
```

```python
#Viaje por el rio - Programación dinámica
################################################################################

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#999 se puede sustituir por float("inf")


#Calculo de la matriz de PRECIOS y RUTAS
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################

PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

```python
import random

#Encontrar 2 puntos mas cercanos
################################################################################

#Lista de numeros
semilla = 42
random.seed(semilla)

#Vector de 1D
VectorD1 = random.sample(range(1000, 9999), 100)


#Algoritmo por fuerza bruta
################################################################################
def fuerza_bruta_puntos_cercanos(vector):

    n = len(vector)

    # Si no hay al menos 2 puntos, se retornara vacio ya que no hay una distancia que se pueda calcular
    if n < 2:
        return None

    #Inicializar las variables
    distancia_minima = float('inf')
    punto1, punto2 = None, None

    #Crear un diccionario con la solución
    solucion = { 'distancia_minima': distancia_minima, 'punto1': punto1, 'punto2': punto2 }

    [solucion.update({'distancia_minima': abs(vector[i] - vector[j]),
                      'punto1': vector[i],
                      'punto2': vector[j]})
    for i in range(n - 1) for j in range(i + 1, n) if abs(vector[i] - vector[j]) < solucion['distancia_minima']]

    return solucion

#Encontrar la solución
solucion = fuerza_bruta_puntos_cercanos(VectorD1)
print(solucion)

#Complejid
# 1) El primer ciclo se ejecuta n-1 veces
# 2) El segundo ciclo se ejecuta n-i-1 veces por cada iteración del primer ciclo
# O(n) = (n-1) + (n-2) + (n-3) + ... + 1 = n(n-1)/2
# O(100) = 4950
```

```python
def distancia_minima_divide_y_venceras(vector):

  n = len(vector)

  # No hay suficientes puntos para calcular la distancia mínima
  if n < 2:
        return None

  # Ordena los puntos en el espacio unidimensional
  vector.sort()

  return divide_y_venceras(vector, 0, n - 1)

def divide_y_venceras(vector, inicio, fin):

    if inicio == fin:
        return None  # No hay suficientes puntos para calcular la distancia mínima

    mitad = (inicio + fin) // 2

    # Divide el problema en dos subproblemas
    izquierda = divide_y_venceras(vector, inicio, mitad)
    derecha = divide_y_venceras(vector, mitad + 1, fin)

    # Combina las soluciones de los subproblemas
    min_izquierda = abs(vector[mitad] - vector[inicio])
    min_derecha = abs(vector[fin] - vector[mitad])
    distancia_minima = min(min_izquierda, min_derecha)

    # Considera el caso de que el par de puntos con la distancia mínima cruce la línea de división
    min_entre_division = abs(vector[mitad + 1] - vector[mitad])
    distancia_minima = min(distancia_minima, min_entre_division)

    # Determina qué par de puntos tiene la distancia mínima
    if min_izquierda < min_derecha and min_izquierda < min_entre_division:
        return {'puntos': (vector[inicio], vector[mitad]), 'distancia': min_izquierda}
    elif min_derecha < min_entre_division:
        return {'puntos': (vector[mitad], vector[fin]), 'distancia': min_derecha}
    else:
        return {'puntos': (vector[mitad], vector[mitad + 1]), 'distancia': min_entre_division}

resultado = distancia_minima_divide_y_venceras(VectorD1)
print("Par de puntos con distancia mínima:", resultado)

#La complejidad es O(n) = log2(n)
#O(n) = 6,64 aunque la solución no es la optima global
```
